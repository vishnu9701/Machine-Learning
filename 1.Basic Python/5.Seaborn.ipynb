# Seaborn
Seaborn is a library that uses Matplotlib underneath to plot graphs. It will be used to visualize random distributions.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

## sns.scatterplot()
This is used to create scatter plots, we can pass data for the x and y-axis and itâ€™ll create a scatter plot for them.

```python
sns.scatterplot(x = df['math score'], y = df['reading score'])
plt.show()
```

```python
plt.figure(figsize = (9,6))
sns.scatterplot(x = 'math score', 
                y = 'reading score',
                hue = 'gender', 
                data = df,
                alpha = 0.8
)
plt.show()
```

## sns.countplot()
This creates a frequency plot that tells the no. of occurrence of the corresponding element in the column.

```python
plt.figure(figsize = (9,6))
sns.countplot(x = 'race/ethnicity', 
              data = df
)
plt.show()
```

## sns.displot()
This creates a histogram for the corresponding distribution. By default, it displays the KDE of the distribution which you can turn off by setting kde = False.

```python
plt.figure(figsize = (9,6))
sns.displot(x = df['math score'], kde = False)
plt.show()
```

## sns.kdeplot()
This creates a Kernel Density Estimate plot or KDE plot of the distribution

```python
plt.figure(figsize = (9,6))
sns.kdeplot(x = df['math score'])
plt.show()
```

##  Histrogram

```python
sns.histplot(df["age"],bins=15,kde=True)
```

## sns.regplot()
This creates a scatter plot that contains a regression line for the data. Weâ€™ll learn about regression lines in the next article but for now, think of it as a line that tells the relationship between 2 points.

```python
plt.figure(figsize = (9,6))
sns.regplot(x = df['math score'], 
            y = df['reading score'],
            scatter_kws = {'color':'pink'},
            line_kws = {'color':'red'}
)
plt.show()
```

## sns.lmplot()
This is almost the same as regplot but it can create a regression line for all the categories of columns set as hue.

```python
sns.lmplot(x = 'math score', 
           y = 'reading score',
           hue = 'gender',
           data = df
)
plt.show()
```

## sns.pairplot()
This plots pairwise bivariate distributions in a dataset. pairplot() only works for numerical columns. You have to pass the subsetted data with columns you want to plot distribution in between.

```python
sns.pairplot(df[['math score',
                 'reading score',
                 'writing score']]
)
plt.show()
```

## sns.jointplot()
This plots both univariate and bivariate distribution between 2 columns.

```python
sns.jointplot(x = 'math score', 
              y = 'reading score',
              data = df
)
plt.show()
```

## sns.boxplot()
Creates a boxplot that gives 5 number summary of the distribution. You can create multiple box plots too. Boxplots also display the outliers in the dataset.

```python
sns.boxplot(data = df)
plt.show()
```

# sns.heatmap()
Plots heatmap of the data. Most commonly used to create correlation heatmaps. annot = True is used to display the value of the cell

```python
sns.heatmap(df.corr(),
            annot = True,
            cmap = 'inferno')
plt.show()
```
